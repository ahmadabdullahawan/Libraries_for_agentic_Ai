```python
#directory loader
from langchain_community.document_loaders import PyPDFLoader,DirectoryLoader
```

```python

loader = DirectoryLoader(path="directory", glob="*.pdf", loader_cls=PyPDFLoader)
doc = loader.load()
print(len(doc))
```

```python
loader = DirectoryLoader(path="directory", glob="*.pdf", loader_cls=PyPDFLoader)
doc = loader.load()
for i in doc:
    print(i.metadata)
```

```python
loader = DirectoryLoader(path="directory", glob="*.pdf", loader_cls=PyPDFLoader)
doc = loader.lazy_load()
for i in doc:
    print(i.metadata)
```
