```python
from langchain_community.document_loaders import TextLoader
```

```python
loader=TextLoader("sample4.txt")
doc=loader.load()
print(len(doc))
```

```python
doc[0].page_content
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=30)
splitted=splitter.split_documents(doc)
print(len(splitted))
```

```python
splitted[0].page_content
```

```python
!pip install faiss-cpu
```

```python
#creating embeddings and STORING IN FAISS 
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
embedding=OllamaEmbeddings(model="qwen3-embedding:0.6b")
```

```python
db=FAISS.from_documents(splitted,embedding)
```

```python
query="How does FAISS vector search work?"
search=db.similarity_search(query)
print(len(search))
```

```python
print(search[0].page_content)
```

```python
retriever=db.as_retriever()
result=retriever.invoke(query)
```

```python
result
```

```python
#embeddig query and then doing similarity search
embedded_query=embedding.embed_query(query)
search=db.similarity_search_by_vector(embedded_query)
```

```python
search
```

```python
query="How does FAISS vector search work?"#similarity search with scores
search=db.similarity_search_with_score(query)
search
```

```python
#saving db 
db1=db.save_local('Faiss_index')
db1
```
