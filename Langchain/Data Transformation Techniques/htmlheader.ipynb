```python
#html header text splittor 
#used to chunk HTML documents based on their structural headings
#(<h1>, <h2>, <h3>, etc.) rather than splitting text blindly by character counts.
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html='''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>AI and Machine Learning Overview</title>
</head>
<body>

    <!-- Main Title -->
    <h1>Introduction to Artificial Intelligence</h1>
    <p>Artificial Intelligence (AI) simulates human intelligence processes through computer systems. These processes include learning, reasoning, and self-correction. Today, AI powers everything from search engines to autonomous vehicles.</p>
    <p>As the technology matures, understanding its core pillars becomes essential for developers and businesses alike.</p>

    <!-- Section 1 -->
    <h2>1. Foundations of Machine Learning</h2>
    <p>Machine Learning (ML) is a subset of AI that allows software applications to become more accurate at predicting outcomes without being explicitly programmed. It relies on algorithms that build mathematical models based on sample data.</p>

    <!-- Subsection 1a -->
    <h3>Supervised Learning</h3>
    <p>In supervised learning, the model is trained on a labeled dataset. This means every training example is paired with its correct output label. Common examples include spam detection and housing price prediction.</p>

    <!-- Subsection 1b -->
    <h3>Unsupervised Learning</h3>
    <p>Unsupervised learning models work with unlabeled data. The system attempts to find hidden patterns, structures, or anomalies on its own. It is heavily used for customer segmentation and recommendation engines.</p>

    <!-- Section 2 -->
    <h2>2. Large Language Models (LLMs)</h2>
    <p>Large Language Models represent the latest breakthrough in Natural Language Processing. These models contain billions of parameters and are trained on massive text corpora to predict the next word in a sequence.</p>

    <!-- Subsection 2a -->
    <h3>The Role of Transformers</h3>
    <p>The Transformer architecture changed how machines process language. By using self-attention mechanisms, a Transformer processes entire sentences at once rather than word-by-word, keeping long-range context intact.</p>

    <!-- Section 3 -->
    <h2>3. Challenges in Modern AI</h2>
    <p>Despite rapid advancements, AI systems face severe challenges. Data privacy concerns, algorithmic bias, and high computational costs limit the widespread adoption of massive foundational models.</p>

</body>
</html>
'''
sepreating=[
    ('h1','header1'),
    ('h2','header2'),
    ('h3','header3'),
  
]
splitter=HTMLHeaderTextSplitter(headers_to_split_on=sepreating)
spltting=splitter.split_text(html)
```

```python
print(spltting)
print(len(spltting))
print(spltting[0])
```

```python
url="https://www.geeksforgeeks.org/"
var=[
    ('h1','Header 1'),
    ('h2','Header 2'),
    ('h3','Header 3'),
]
splitter=HTMLHeaderTextSplitter(headers_to_split_on=var)
doc=splitter.split_text_from_url(url)
print(len(doc))
print(doc)
print(doc[0].page_content)
```
