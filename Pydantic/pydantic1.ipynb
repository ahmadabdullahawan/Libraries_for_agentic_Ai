#creating models in pydantic

pydantic models are foundation of data validation in python.They use python type annotation to define structure and validate data at run time.

#Type annotation?

allow you to explicitly declare the expected data types of variables, function arguments, and return values

```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str
    
person0=Person(name="abdullah",age=23,city="kohat")
print(person0)
```

```python
person1=Person(name="Ali",age=23,city=23)
```

```python
#comparing data class with BaseModel 
#data class 
#A shortcut for creating standard Python classes that exist mostly just to hold data
#it saves you from writing boring "boilerplate" code like below
```

```python
# 🛑 The Old, Tedious Way
class Movie:
    def __init__(self, title: str, year: int):
        self.title = title
        self.year = year
```

```python
#  The Data Class Way
from dataclasses import dataclass

@dataclass
class Movie:
    title: str
    year: int
```

```python
#model with optional fields
from typing import Optional 
class Person(BaseModel):
    name:str
    age:int
    salary:Optional[float]=None
    is_strudent:Optional[bool]=True
    
p0=Person(name="abdullah",age=23)
print(p0)
```

```python
p1=Person(name="Ali",age=23,salary='1000',is_student=False)
print(p1)#se the output how pydantic handles "1000"
```

```python
from typing import List
class classroom(BaseModel):
    students:list[str]
    age:list[int]
    
c1=classroom(students=["abdullah","hashir"],age=[23,23])
print(c1)
```

```python
#handling errors
from typing import Optional 
class Person(BaseModel):
    name:str
    age:int
    salary:Optional[float]=None
    is_strudent:Optional[bool]=True
    
p0=Person(name="abdullah",age=23)
print(p0) 
```

```python
try:
    p1=Person(name=123,age=12)
except ValueError as e:
    print(e)
    print("Validation error")
```

```python
# nested modeling
from pydantic import BaseModel
class Address(BaseModel):
    city:str
    zip_code:int
    
class customer(BaseModel):
    customer_id:int
    customer_name:str
    address: Address
    
cust1=customer(customer_id=1,customer_name="Abdullah",address={"city":"Kohat","zip_code":"1234"})
print(cust1)
        
```

```python
#pydantic fields 
from pydantic import BaseModel,Field
class Person(BaseModel):
    name:str=Field(min_length=2,max_length=50)
    age:int=Field(ge=0,le=30) #greaterthanequalto 0 and lessthanequalto30
    
p1=Person(name='Abdullah',age=18)
print(p1)
```

```python
p2=Person(name='Abdullah',age=40)
print(p2)
```

```python
from pydantic import BaseModel, Field

class User(BaseModel):
    username: str = Field(..., description="Unique username for the user")
    age: int = Field(default=18, description="User age, defaults to 18")
    email: str = Field(default_factory=lambda: "user@example.com", description="Default email address")

# Examples
user1 = User(username="alice")
print(user1)  # username='alice' age=18 email='user@example.com'

user2 = User(username="bob", age=25, email="bob@domain.com")
print(user2)  # username='bob' age=25 email='bob@domain.com'
```
